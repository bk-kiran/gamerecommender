Feature Selection Part

```python
games.columns
```

```python
games=games[['Unnamed: 0', 'Title', 'Summary', 'Genres']]
```

```python
games
```

```python
games['tags'] = games['Summary']+games['Genres']
```

```python
games
```

```python
new_data = games.drop(columns=['Summary', 'Genres'])
```

```python
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=1512, stop_words='english')
```

```python
cv
```

```python
vector=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['Title']=="Hades"].index[0]
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[3:10]:
    print(new_data.iloc[i[0]].Title)
```

```python
def recommend(games):
    index = new_data[new_data['Title']==games].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    list_of_games = []
    for i in distance[3:12]:
        check = new_data.iloc[i[0]].Title
        if check not in list_of_games:
            list_of_games.append(check)

    for j in list_of_games:
        print(j)
    
        
```

```python
recommend('The Legend of Zelda: Breath of the Wild')
```
